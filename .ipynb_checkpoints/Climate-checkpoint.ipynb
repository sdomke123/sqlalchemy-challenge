```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

## Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
inspector = inspect(engine)
inspector.get_table_names()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## Exploratory Precipitation Anlysis

```python
# Find the most recent date in the data set.
most_recent = session.query(Measurement.date).order_by(Measurement.date.desc()).first().date
most_recent
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database.
# Calculate the date one year from the last date in data set.
# Perform a query to retrieve the data and precipitation scores
past_year_data = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= '2016-08-24').\
                 filter(Measurement.date <= '2017-08-23').\
                 filter(Measurement.station == 'USC00516128').group_by(Measurement.date).all()
past_year_data
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
# Sort the dataframe by date
past_year_df = pd.DataFrame(past_year_data, columns = ["Date", "Precipitation (in)"])
past_year_df.sort_values(by=["Date"])
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
past_year_df.plot()
plt.xlabel("Date (by ID)")
plt.ylabel("Precipitation (in)")
plt.title("Last 12 Months of Precipitation Recorded in Hawaii")
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
past_year_df.describe()
```

## Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
session.query(Station.id).count()
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
MS = Measurement.station
station_counts = session.query(MS, func.count(MS)).group_by(MS).order_by(func.count(MS).desc()).all()
station_counts
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
MT = Measurement.tobs
#most_active = station_counts[0][0]
session.query(func.min(MT), func.max(MT), func.avg(MT)).filter(Measurement.station == station_counts[0][0]).all()
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station.
best_past_year_temps = session.query(Measurement.station, Measurement.tobs).filter(Measurement.date >= '2016-08-24').\
                 filter(Measurement.date <= '2017-08-23').\
                 filter(Measurement.station == station_counts[0][0]).all()
best_past_year_temps
```

```python
#Plot the results as a histogram.
bpyt_df = pd.DataFrame(best_past_year_temps)
bpyt_df.plot.hist(bins=12)
plt.xlabel("Temperature")
plt.title("Last 12 Months Worth of Precipitation Recorded by Station USC00519281")
```

## Close session

```python
# Close Session
session.close()
```
